Q1. How many requests that were sent on '30/Apr/1998' had an HTTP return code 404?

The above requested information can be fetched from the data provided in the CSV by firstly grouping the data based on Date and the ResponseCode. Next, we just need to fetch the respective group (in this case, the one pertaining to April 30, 1998 and HTTP code 404)

```python
log_df = pd.read_csv("wc_day6_1_sample.csv",
                     names=['ClientID', 'Date', 'Time', 'URL', 'ResponseCode', 'Size'],
                     na_values=['-'])

multi_grouped = log_df.groupby(['Date', 'ResponseCode']).get_group(('30/Apr/1998',404))
print len(multi_grouped)
```

Q2. Generate a plot of the number of distinct users of the site every hour on 01/May/1998 using bar chart.

Steps taken to achieve the above:
- Filter entries for the requested date
- Convert all the time strings to timestamps and then to their respective hours (easier to plot)
- Group all the data depending on their respective hour numbers
- Get length of unique users for the particular hour and plot using a line plot.
    

```python
may_01_df = log_df[log_df['Date'] == '01/May/1998']

may_01_df_grouped_by_time = may_01_df.groupby(lambda x: pd.to_datetime(may_01_df['Time'][x]).hour)
unique_users_per_hour = [len(may_01_df_grouped_by_time.get_group(item).groupby('ClientID')) for item in may_01_df_grouped_by_time.groups]
plot_01_data = {'NumUniqueUsers':unique_users_per_hour}
plot_01_df = pd.DataFrame(plot_01_data)
plot_01 = plot_01_df.plot(y = 'NumUniqueUsers', x = plot_01_df.index, kind='bar', ylim=[min(unique_users_per_hour), max(unique_users_per_hour)])
```

Q3. Generate a line plot of the number of distinct users vs size of traffic of the site every hour on 01/May/1998, is there any correlation? (hint: use "hour_grouped['ClientID'].nunique()" to get the number of distinct users)

Steps taken to achieve the above:
- Similar to unique users per hour, find out site traffic for the particular hour
- use subplots to plot unique users and site traffic per hour (in both, the x axis represents the repsective hour)
- The resulting graph clearly shows a positive correlation between the two curves

```python
site_traffic_per_hour = [len(may_01_df_grouped_by_time.get_group(item).groupby('Size')) for item in may_01_df_grouped_by_time.groups]

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(range(len(unique_users_per_hour)), unique_users_per_hour, 'g-')
ax2.plot(range(len(site_traffic_per_hour)), site_traffic_per_hour, 'r-')

ax1.set_xlabel('Hour of the day')
ax1.set_ylabel('# Requests', color='g')
ax2.set_ylabel('Size of traffic handled', color='r')
```

Q4. We wish to see if there is any correlation between client-ids and hours of the day at which they visit the website. Get 100 client ids from the dataset and generate a scatter plot that shows the hours of the day these clients sent requests. Hint: df.plot(kind='scatter', x='a', y='b'); and df['Column'].unique()

Steps taken to achieve above:
- Get a set of all the unique client IDs and get the top 100 unique IDs
- Get a condensed set of entries having client IDs belonging to the 100 unique IDs
- Convert the time string in the condensed set to the respective hour
- plot ClientID on x axis and time on y axis to try to find a correlation between the two.
- The scatter plot generated below clearly shows that with the increase in value of ClientID, the range of hours they access the site reduces (higher range of hour values for IDs < 5000)

```python
test_set = (log_df['ClientID'].unique())[:100]
condensed_set = log_df[log_df['ClientID'].isin(test_set)]
condensed_set['Time'] = [pd.to_datetime(condensed_set['Time'][i]).hour for i in condensed_set.index.values]
xlimForPlot02 = [min(condensed_set['ClientID']), max(condensed_set['ClientID'])]
ylimForPlot02 = [min(condensed_set['Time']), max(condensed_set['Time'])]
plot_02 = condensed_set.plot(x = 'ClientID', y = 'Time', kind='scatter', xlim = xlimForPlot02, ylim = ylimForPlot02)
```
